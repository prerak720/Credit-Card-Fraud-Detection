```python
from sklearn.preprocessing import StandardScaler
```

Normalize the data by using a stand-alone scaler. Why? Normalization is used to assign all values to a scale between 1 and 0, which prevents bias, speeds processing, and reduces overfitting.

```python
s = StandardScaler()
```

```python
df['Amount'] = s.fit_transform(df['Amount'].values.reshape(-1, 1))
# transforming the data
```

```python
df = df.drop(columns=['Time'])
#drop the time column beacuse dont use for prediction
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
#split the data 
```

```python
x = df.iloc[:,:-1]
y = df[['Class']]
# x is independent column 
# y is dependent column
```

```python
x
```

```python
y
```

```python

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=22)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo = LogisticRegression()
```

```python
model = lo.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(prediction,y_test)
```

```python
confusion_matrix(prediction,y_test)
```

```python
# logical regression get 99% accurancy 
```
